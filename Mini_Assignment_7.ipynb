##Milti-Level Indexing

1. Read the [population](https://datascience.quantecon.org/assets/data/wdi_population.csv) dataset, [financial](https://datascience.quantecon.org/assets/data/wdi_data.csv) dataset, and [area](https://drive.google.com/file/d/1fY1ee-Uqsu7Bw54GbT4L0m5dCwTttksG/view?usp=sharing) dataset separately as dataframes. For the first two, make sure to make 'country' and 'year' columns as first and second level index respectively.

```
import pandas as pd
import numpy as np
pop = pd.read_csv('wdi_population.csv', na_values=['na','n/a'])
pop = pop.set_index(['country', 'year'])
pop
```

```
fin = pd.read_csv('wdi_data.csv', na_values=['na','n/a'])
fin = fin.set_index(['country', 'year'])
fin
```

```
area = pd.read_csv('Area Size.csv', na_values=['na','n/a'])
area
```

2. From the financial dataset, display the three columns of 'GovExpend', 'Exports', and 'Imports' for Canada over all the years. Assign the output to a variable called 'findf_canada'.

```
findf_canada = fin.loc[('Canada',slice(None)), ['GovExpend', 'Exports', 'Imports']]
findf_canada
```

3. From the financial dataset, display the three columns of 'GovExpend', 'Exports', and 'Imports' for all the countries over 2017. Store the output in a variable called 'findf_2017'.

```
findf_2017 = fin.loc[(slice(None),2017), ['GovExpend', 'Exports', 'Imports']]
findf_2017
```

##Concatenating Data

4. Concatenate the two financial and population dataframes into a new dataframe that includes both financial and population data for each country and each year. Call the new dataset 'finpop'.

```
finpop = pd.concat([fin,pop])
finpop
```

##Merging Data

5. Reset the indexes for the finpop and findf_2017 datasets to the default index numbers using [.reset_index()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.reset_index.html) method. Make sure to store them back in their own dataframe after reseting teh index.

```
finpop = finpop.reset_index()

findf_2017 = findf_2017.reset_index()
```

6. Merge the finpop and area size datasets, so the output includes all the columns from both datasets for each country each year. Make sure it includes all the countries available in the area size dataframe (including Russia).

```
pd.merge(finpop,area, on='country', how = 'outer')
```

7. Merge the findf_2017 and the area size datasets. Make sure the output includes all the countries available in either of the two datasets (including Russia).

```
pd.merge(findf_2017,area, on ='country', how = 'outer')
```
