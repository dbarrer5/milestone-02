```
import pandas as pd
import matplotlib.pyplot as plt
```

2. Create a slice of financial dataset for the year 2017 rows, and create another slice for the year 2000 rows; store them in separate variables called 'fin2017' and 'fin2000' respectively.

 Use the above two variables to creae ONE line plot with two lines; one displays the GDP of the countries for 2017 and the other one displays the GDP of the countries for 2000. Make sure to:

*   set the x axis as country names and y axis as GDP
*   set proper linestyle and marker for each line
*   set proper color for line and markers
*   set proper legend for the plot with labels for each line
*   set proper plot title and axes labels


```
pop = pd.read_csv('wdi_population.csv')
fin = pd.read_csv('wdi_data.csv')
```

```
fin2017 = fin[fin.year==2017]
fin2000 = fin[fin.year==2000]
fin2017
fin2000

plt.plot(fin2017.country, fin2017.GDP, ls = "-", label = "2017",color="blue", marker = "o", markerfacecolor="blue")
plt.plot(fin2000.country, fin2000.GDP, ls = "-.", label = "2000",color="red", marker = "v", markerfacecolor="red")
plt.legend()
plt.xlabel("Country")
plt.ylabel("GDP")
plt.title("GPA per Country")
plt.show()
```

4. Slice the financial dataset for the rows for the United States. Then, use that to create a scatter plot to display the 'Exports' against the 'Imports' of the United States, and use the 'year' as color for the markers; it means the markers (data points) of later years will be shown with brighter colors on the scatterplot.

```
fin_us = fin[fin.country== "United States"]
```

```
plt.scatter("Exports", "Imports", data = fin_us, c = 'year')
plt.xlabel("Exports")
plt.ylabel("Imports")
plt.title("Exports and Imports for US")
plt.show()
```
