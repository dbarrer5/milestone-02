##Data Ingestion & Cleaning


1. Ingest [the Census 2017 dataset](https://drive.google.com/file/d/1bXZmQVLq0bG29mmIhsQneXUzCQ20Qpaj/view?usp=sharing) into a dataframe. Make sure string like ‘na’ and ‘NA’ are identified as missing values. Then check out the first few rows, descriptive statistics, and data type for each column of the dataset.

```
import pandas as pd
import numpy as np

census = pd.read_csv('cns2017_county_data.csv', na_values=['na','n/a'])
census.head() #first few rows
census.describe() # descriptive statistics
census.info() #data tpye for each column or census.dtypes
```

2. Check if there are any duplicate records (rows). How many duplicate records are there?

```
#The number of True can be counted with sum() method.
census.duplicated().sum()
```

3. Display the ratio (in percentage) of missing values for each column. Then drop any column that has over 50% of missing values.

 Next, do the same for the rows; meaning drop the rows with over 50% missing values. Display the number of rows before and after you drop such rows.

```
((census.isnull().sum()/len(census)*100).sort_values())

#No columns with 50% missing values
census = census.dropna(axis = 1, thresh = len(census)*0.5)
#Drop rows with over 50%
census = census.dropna(axis = 0, thresh = 37*0.5)
len(census.index)
census
```

4. For the remaining missing values in the dataframe, fill them all with the median value of its column.

```
#Did not include columns with 0% missing values
census = census.fillna(census.median())
census
```

## Data Preparation

5. Encode the state names using label encoding, and store the encoded column in a new column called 'state_coded'.

```
census["State"] = census["State"].astype('category')
census["state_coded"] = census["State"].cat.codes
census
```

6. Bin the TotalPop column into six equally-sized bins, and store it as a new column called "TotalPop_bins".

```
bins = pd.qcut(census['TotalPop'],q=6)

census['TotalPop_bins'] = bins
census
```
